## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**


```python
class SistemaNotas:
    def __init__(self, nome, disciplina, nota):
        self.nome = nome
        self.disciplina = disciplina
        self.nota = nota

    def exibir_dados(self):
        print(f"Aluno: {self.nome}")
        print(f"Disciplina: {self.disciplina}")
        print(f"Nota: {self.nota}")

    def verificar_aprovacao(self):
        if self.nota >= 7:
            return "Aprovado"
        else:
            return "Reprovado"

    def atualizar_nota(self, nova_nota):
        self.nota = nova_nota

class AlunoOnline(SistemaNotas):
    def __init__(self, nome, disciplina, nota, plataforma):
        super().__init__(nome, disciplina, nota)
        self.plataforma = plataforma

    def exibir_plataforma(self):
        print(f"Plataforma de estudo: {self.plataforma}")

aluno1 = AlunoOnline("Lucas", "Matem√°tica", 9.0, "Google Classroom")
aluno2 = AlunoOnline("Fernanda", "Hist√≥ria", 6.5, "Moodle")

aluno1.exibir_dados()
aluno1.exibir_plataforma()
print(aluno1.verificar_aprovacao())

print()

aluno2.exibir_dados()
aluno2.exibir_plataforma()
print(aluno2.verificar_aprovacao())
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class SistemaNotas:
    def __init__(self, nome, disciplina, nota):
        self.nome = nome
        self.disciplina = disciplina
        self.__nota = nota  # Atributo privado

    def exibir_dados(self):
        print(f"Aluno: {self.nome}")
        print(f"Disciplina: {self.disciplina}")
        print(f"Nota: {self.__nota}")

    def verificar_aprovacao(self):
        if self.__nota >= 7:
            return "Aprovado"
        else:
            return "Reprovado"

    def atualizar_nota(self, nova_nota):
        if 0 <= nova_nota <= 10:  # L√≥gica de valida√ß√£o
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")

    def get_nota(self):
        return self.__nota  # M√©todo 'get' para acessar a nota

    def set_nota(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota  # M√©todo 'set' para modificar a nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")


class AlunoOnline(SistemaNotas):
    def __init__(self, nome, disciplina, nota, plataforma):
        super().__init__(nome, disciplina, nota)
        self.plataforma = plataforma

    def exibir_plataforma(self):
        print(f"Plataforma de estudo: {self.plataforma}")


aluno1 = AlunoOnline("Lucas", "Matem√°tica", 9.0, "Google Classroom")
aluno2 = AlunoOnline("Fernanda", "Hist√≥ria", 6.5, "Moodle")

# Exibindo dados
aluno1.exibir_dados()
aluno1.exibir_plataforma()
print(aluno1.verificar_aprovacao())

print()

aluno2.exibir_dados()
aluno2.exibir_plataforma()
print(aluno2.verificar_aprovacao())

# Usando os m√©todos get e set
print(f"Nota atual do aluno1: {aluno1.get_nota()}")
aluno1.set_nota(8.5)  # Atualizando a nota
print(f"Nota atualizada do aluno1: {aluno1.get_nota()}")

aluno2.set_nota(11)  # Tentando definir uma nota inv√°lida
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

class SistemaNotas:
    def __init__(self, nome, disciplina, nota):
        self.nome = nome
        self.disciplina = disciplina
        self.__nota = nota

    def exibir_dados(self):
        print(f"Aluno: {self.nome}")
        print(f"Disciplina: {self.disciplina}")
        print(f"Nota: {self.__nota}")

    def verificar_aprovacao(self):
        if self.__nota >= 7:
            return "Aprovado"
        else:
            return "Reprovado"

    def atualizar_nota(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")

    def get_nota(self):
        return self.__nota

    def set_nota(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")

    def inserir_bd(self):
        pass  # M√©todo que ser√° sobrescrito pela subclasse

class AlunoOnline(SistemaNotas):
    def __init__(self, nome, disciplina, nota, plataforma):
        super().__init__(nome, disciplina, nota)
        self.plataforma = plataforma

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Plataforma de estudo: {self.plataforma}")

    def inserir_bd(self):
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS alunos (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                disciplina TEXT,
                nota REAL,
                plataforma TEXT
            )
        ''')
        cursor.execute('''
            INSERT INTO alunos (nome, disciplina, nota, plataforma)
            VALUES (?, ?, ?, ?)
        ''', (self.nome, self.disciplina, self.get_nota(), self.plataforma))
        conn.commit()
        conn.close()

    def excluir_bd(self):
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('''
            DELETE FROM alunos WHERE nome = ? AND disciplina = ?
        ''', (self.nome, self.disciplina))
        conn.commit()
        conn.close()

    @staticmethod
    def consultar_bd():
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM alunos')
        alunos = cursor.fetchall()
        conn.close()
        return alunos

# Testando as funcionalidades
aluno1 = AlunoOnline("Lucas", "Matem√°tica", 9.0, "Google Classroom")
aluno1.exibir_dados()

# Inserindo no banco de dados
aluno1.inserir_bd()

# Consultando o banco de dados
alunos = AlunoOnline.consultar_bd()
print("\nAlunos registrados no banco de dados:")
for aluno in alunos:
    print(aluno)

# Excluindo o aluno do banco de dados
aluno1.excluir_bd()
print("\nAp√≥s excluir o aluno:")
alunos = AlunoOnline.consultar_bd()
for aluno in alunos:
    print(aluno)
```

```python
import sqlite3

# Classe principal que representa o sistema de notas
class SistemaNotas:
    def __init__(self, nome, disciplina, nota):
        """
        Construtor da classe principal. Inicializa os atributos b√°sicos do aluno.
        """
        self.nome = nome
        self.disciplina = disciplina
        self.__nota = nota  # Atributo privado

    def exibir_dados(self):
        """
        Exibe os dados do aluno: nome, disciplina e nota.
        """
        print(f"Aluno: {self.nome}")
        print(f"Disciplina: {self.disciplina}")
        print(f"Nota: {self.__nota}")

    def verificar_aprovacao(self):
        """
        Verifica se o aluno est√° aprovado ou reprovado com base na nota.
        """
        if self.__nota >= 7:
            return "Aprovado"
        else:
            return "Reprovado"

    def atualizar_nota(self, nova_nota):
        """
        Atualiza a nota do aluno, se a nova nota for v√°lida (entre 0 e 10).
        """
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")

    def get_nota(self):
        """
        Retorna a nota do aluno.
        """
        return self.__nota

    def set_nota(self, nova_nota):
        """
        Define uma nova nota para o aluno, se for v√°lida.
        """
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida. A nota deve estar entre 0 e 10.")

    def inserir_bd(self):
        """
        M√©todo que ser√° sobrescrito pela subclasse. Colocado aqui para manter a estrutura.
        """
        pass

# Subclasse que herda de SistemaNotas, representando um aluno online
class AlunoOnline(SistemaNotas):
    def __init__(self, nome, disciplina, nota, plataforma):
        """
        Construtor da subclasse. Inicializa os atributos do aluno e a plataforma de estudo.
        """
        super().__init__(nome, disciplina, nota)
        self.plataforma = plataforma

    def exibir_dados(self):
        """
        Sobrescreve o m√©todo exibir_dados da classe principal para incluir a plataforma de estudo.
        """
        super().exibir_dados()
        print(f"Plataforma de estudo: {self.plataforma}")

    def inserir_bd(self):
        """
        Insere o aluno no banco de dados SQLite.
        """
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS alunos (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                disciplina TEXT,
                nota REAL,
                plataforma TEXT
            )
        ''')
        cursor.execute('''
            INSERT INTO alunos (nome, disciplina, nota, plataforma)
            VALUES (?, ?, ?, ?)
        ''', (self.nome, self.disciplina, self.get_nota(), self.plataforma))
        conn.commit()
        conn.close()

    def excluir_bd(self):
        """
        Exclui o aluno do banco de dados com base no nome e disciplina.
        """
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('''
            DELETE FROM alunos WHERE nome = ? AND disciplina = ?
        ''', (self.nome, self.disciplina))
        conn.commit()
        conn.close()

    @staticmethod
    def consultar_bd():
        """
        Consulta todos os alunos no banco de dados.
        """
        conn = sqlite3.connect('sistema_notas.db')
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM alunos')
        alunos = cursor.fetchall()
        conn.close()
        return alunos

# Testando as funcionalidades do c√≥digo

# Criando inst√¢ncias de alunos
aluno1 = AlunoOnline("Lucas", "Matem√°tica", 9.0, "Google Classroom")
aluno2 = AlunoOnline("Fernanda", "Hist√≥ria", 6.5, "Moodle")

# Exibindo dados dos alunos
print("Exibindo dados do aluno 1:")
aluno1.exibir_dados()
print(f"Aprovado: {aluno1.verificar_aprovacao()}")
print()

print("Exibindo dados do aluno 2:")
aluno2.exibir_dados()
print(f"Aprovado: {aluno2.verificar_aprovacao()}")
print()

# Atualizando nota de um aluno
aluno1.set_nota(8.5)
print(f"Nota atualizada do aluno 1: {aluno1.get_nota()}")
print()

# Inserindo alunos no banco de dados
aluno1.inserir_bd()
aluno2.inserir_bd()

# Consultando alunos no banco de dados
print("Alunos registrados no banco de dados:")
alunos = AlunoOnline.consultar_bd()
for aluno in alunos:
    print(aluno)

# Excluindo o aluno 1 do banco de dados
aluno1.excluir_bd()
print("\nAp√≥s excluir o aluno 1:")
alunos = AlunoOnline.consultar_bd()
for aluno in alunos:
    print(aluno)
```
